```python
#reading the .csv file into a dataframe
df = pd.read_csv('/content/gdrive/MyDrive/Task2hp/HPSpells.csv')
```

```python
df.head()
```

```python
df.drop(df[df.Type=='Hex'].index,inplace=True)
```

```python
df.head()
```

```python
df.dropna(subset=['Light'],inplace=True)
```

```python
df.head()
```

```python
freq ={}
for Type in df.Type.unique():
  c=0
  for item in df.Type:
    if item==Type:
       c+=1
  freq[Type]=c

plt.figure(figsize=(10,7))
plt.xlabel('Spells')
plt.ylabel('Number of Spells')
plt.title('Spell distribution')
plt.barh(list(freq.keys()),list(freq.values()),color='green')
plt.show()
```
